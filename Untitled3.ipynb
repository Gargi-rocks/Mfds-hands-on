```python
import matplotlib.pyplot as plt
url="https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
var=pd.read_csv(url)
var['Fare'].plot(kind='bar', title='Fare distribution')
plt.xlabel('money')
plt.show()

```

```python
import matplotlib.pyplot as plt
url="https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
var=pd.read_csv(url)
var['Age'].value_counts().plot(kind='hist', title='Age Distribution')
plt.xlabel('Ages')
plt.show()
```

```python
import matplotlib.pyplot as plt
url="https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
var=pd.read_csv(url)
var.groupby('SibSp')['Age'].mean().plot(kind='hist', title='Age Distribution')
plt.ylabel('Ages')
plt.show()
```
